```python
n_trials_total = 30
n_trials_per_net = 1
reps = 4
burn_in_time = 300.0
n_procs = 32
record_vsec = False
return_spike_rates = True

sample_dev_peak_rate = list()
sample_dev_ampl = list()
sample_ensemble = list()

# NB: this is the only arg for splot_dev_spiking_v2() that changes between
# the various sim conditions
drive_strengths_all = list()
dev_amplitudes = [-1, 0, 1]

for seed_idx, seed in enumerate(range(n_trials_total)):
    for dev_ampl_idx, dev_ampl in enumerate(dev_amplitudes):
        net, drive_params = sim_dev_spiking(dev_magnitude=dev_ampl,
                                            reps=reps,
                                            n_trials=n_trials_per_net,
                                            burn_in_time=burn_in_time,
                                            n_procs=n_procs,
                                            record_vsec=record_vsec,
                                            rng=np.random.default_rng(seed))

        rep_start_times = drive_params['rep_times']
        drive_times = drive_params['drive_times']
        drive_strengths = drive_params['drive_strengths']
        tstop = drive_params['tstop']

        spiking_fig, spike_rates, dev_rate_peaks = plot_dev_spiking_v2(
            net,
            burn_in_time,
            rep_start_times,
            drive_times,
            drive_strengths,
            tstop,
            return_spike_rates=return_spike_rates
        )

        ensemble_names = ['agg. (eP+eNP)', 'P', 'NP']
        cell_keys = ['L2/3e', 'L2e_1', 'L2e_2']

        # save trial-specific dev peak rate, drive ampl, and
        # mapping to the specific ensemble being measured for
        # plotting latter (used in DataFrame)
        for ens_idx, ensemble_name in enumerate(ensemble_names):
            cell_key = cell_keys[ens_idx]
            sample_dev_peak_rate.append(dev_rate_peaks[cell_key])
            sample_dev_ampl.append(dev_amplitudes[dev_ampl_idx])
            sample_ensemble.append(ensemble_name)

        # create a single networks with aggregated spike trials from
        # all other networks
        if seed_idx == 0:
            if dev_ampl == -1:
                net_neg = net
            elif dev_ampl == 0:
                net_zero = net
            elif dev_ampl == 1:
                net_pos = net
            drive_strengths_all.append(drive_strengths)

        else:
            new_spike_times = net.cell_response._spike_times
            new_spike_gids = net.cell_response._spike_gids
            new_spike_types = net.cell_response._spike_types

            if dev_ampl == -1:
                net_neg.cell_response._spike_times.extend(new_spike_times)
                net_neg.cell_response._spike_gids.extend(new_spike_gids)
                net_neg.cell_response._spike_types.extend(new_spike_types)
            elif dev_ampl == 0:
                net_zero.cell_response._spike_times.extend(new_spike_times)
                net_zero.cell_response._spike_gids.extend(new_spike_gids)
                net_zero.cell_response._spike_types.extend(new_spike_types)
            elif dev_ampl == 1:
                net_pos.cell_response._spike_times.extend(new_spike_times)
                net_pos.cell_response._spike_gids.extend(new_spike_gids)
                net_pos.cell_response._spike_types.extend(new_spike_types)
            
    print(f'completed session {seed_idx}')
```

```python
for net_idx, net in enumerate([net_neg, net_zero, net_pos]):

    spiking_fig, spike_rates, dev_rate_peaks = plot_dev_spiking_v2(
        net,
        burn_in_time,
        rep_start_times,
        drive_times,
        drive_strengths_all[net_idx],
        tstop,
        return_spike_rates=return_spike_rates
    )
    # spiking_fig.savefig(f'detailed_model_dev{dev_amplitudes[net_idx]}.pdf')
```
